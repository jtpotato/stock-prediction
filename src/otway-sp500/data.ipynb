```python
import yfinance as yf

# get S&P 500
data = yf.download("^GSPC", period="10y")

data
```

```python
# Strip Adj Close and Volume
data = data[["Open", "High", "Low", "Close"]]

# add indicators
data["5d"] = data["Close"].rolling(window=5).mean()
data["10d"] = data["Close"].rolling(window=10).mean()
data["20d"] = data["Close"].rolling(window=20).mean()

data.dropna(inplace=True)

data
```

```python
# normalise data
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
# ideally this is pretty similar between runs of data collection, so it doesn't actually matter if we save it or not as long as the model is retrained regularly.

scaled_data.shape
```

```python
# save scaler to file
import joblib

joblib.dump(scaler, "./scaler.save")
```

```python
import pandas as pd
```

```python
# put scaled data back into df
data[["Open", "High", "Low", "Close", "5d", "10d", "20d"]] = scaled_data

data
```

```python
data.to_csv("./sp500.csv")
```
