```python
import pandas as pd
```

```python
stock_list = pd.read_csv("stocks.tsv", sep="\t").sample(n=20, random_state=2024)

stocks = stock_list["STOCK"]

stocks
```

```python
DAYS_LAGGED = 5
COLUMNS = ["Close", "Volume"]
```

```python
training_data = pd.DataFrame()

for stock in stocks:
  stock_dataframe = pd.read_csv(f"tickers/{stock}.csv")
  for i in range(DAYS_LAGGED):
    for column in COLUMNS:
      stock_dataframe[f"{column}-{i + 1}"] = stock_dataframe[column].shift(i + 1)
      stock_dataframe = stock_dataframe.dropna()
  
  training_data = pd.concat([training_data, stock_dataframe], ignore_index=True)

training_data
```

```python
training_data = training_data.drop("Volume", axis="columns")
```

```python
training_data.to_csv("cleaned_data.csv")
```
